```python
import sys
sys.path.append('../')
import mobipy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mobipy.plotting as mplt
from mpl_toolkits.mplot3d import Axes3D
```

```python
app = mobipy.Model_Application.build_from_model_and_data_file('../models/nivafjord_simplycnp_model.txt', 
                                                              '../models/data/NIVAFjord/nivafjord_drammen.dat')
```

```python
# Set parameter values and run the model
app.start_date = '2018-01-01'
app.end_date = '2019-12-01'
app.run()
```

```python
# Extract slices of the temperature and salinity results.
temp, pos, dates = app.layer.water.temp["Drammensfjorden", :18]
sal,  pos, dates = app.layer.water.salinity["Drammensfjorden", :18]
```

```python
fig, (ax0, ax1) = plt.subplots(2, 1, figsize=(8, 8))
mplt.quick_heatmap(ax0, temp, dates, pos)
mplt.quick_heatmap(ax1, sal, dates, pos)
```
